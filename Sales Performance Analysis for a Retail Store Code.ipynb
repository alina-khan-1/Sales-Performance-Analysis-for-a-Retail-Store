1. What does the monthly/yearly sales trend look like? 

### Monthly Sales Trend â€“ Insights
So I plotted the monthly sales data over a 4-year period, and a few interesting patterns really stood out:

- November is consistently the highest month every year (like in 2015, 2016, 2017). This is probably because of big shopping seasons maybe Black Friday, holiday offers, or year-end sales.

- February and April tend to be slower months, with noticeably lower sales. Makes sense could be post-holiday budget crunch or just naturally quiet months.

- Sales in 2017 are clearly higher overall compared to previous years. That suggests the business was growing steadily year by year.

Throughout the chart, there are regular ups and downs, which is pretty normal for any business. You can see those dips and recoveries happening often nothing too alarming, just part of the cycle.

### Conclusion 
I created this line chart to understand how sales have performed over time, month by month. It helps spot both the strong and weak periods. For example, sales are always high around November â€” likely due to seasonal promotions. On the flip side, months like February and April are usually slower. There's also a clear upward trend, especially in 2017, which points to business growth. These kinds of insights are super useful for planning â€” like when to stock more inventory, launch discounts, or run marketing campaigns.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Grouping data by Region
region_group = df_Store_data.groupby('Region')[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False)

# Plotting
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Sales Bar Plot
ax1 = axs[0]
region_group['Sales'].plot(kind='bar', ax=ax1, color='teal')
ax1.set_title('Total Sales by Region')
ax1.set_ylabel('Sales')
ax1.set_xlabel('Region')
for bar in ax1.patches:
    ax1.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height() + 1000,  # Added offset for spacing
        f'{bar.get_height():,.0f}',
        ha='center', va='bottom', fontsize=9
    )

# Profit Bar Plot
ax2 = axs[1]
region_group['Profit'].plot(kind='bar', ax=ax2, color='orange')
ax2.set_title('Total Profit by Region')
ax2.set_ylabel('Profit')
ax2.set_xlabel('Region')
for bar in ax2.patches:
    ax2.annotate(f'{bar.get_height():,.0f}',
                 (bar.get_x() + bar.get_width() / 2, 
                  bar.get_height()), 
                 ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate margin by category
category_margin = df_Store_data.groupby('Category')[['Sales', 'Profit']].sum()
category_margin['Margin (%)'] = (category_margin['Profit'] / category_margin['Sales'] * 100).round(2)

# Calculate margin by region
region_margin = df_Store_data.groupby('Region')[['Sales', 'Profit']].sum()
region_margin['Margin (%)'] = (region_margin['Profit'] / region_margin['Sales'] * 100).round(2)

# Create dual bar chart
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Margin by category
category_margin['Margin (%)'].plot(kind='bar', ax=ax1, color='teal')
for bar in ax1.patches:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width() / 2, 
             height + 0.3, 
             f'{height:.1f}%', 
             ha='center', va='bottom', fontsize=9)

# Margin by region
region_margin['Margin (%)'].plot(kind='bar', ax=ax2, color='teal')
for bar in ax2.patches:
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width() / 2, 
             height + 0.3, 
             f'{height:.1f}%', ha='center', va='bottom', fontsize=9)

# Customize charts
ax1.set_title('Profit Margin by Category')
ax2.set_title('Profit Margin by Region')
ax1.set_xlabel('Category')
ax2.set_xlabel('Region')
ax1.set_ylabel('Margin (%)')
ax2.set_ylabel('Margin (%)')
ax1.grid(True)
ax2.grid(True)
ax1.tick_params(axis='x', rotation=45)
ax2.tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```

#### Conclusion: Highest Profit Margins by Category and Region

The dual bar chart, reveals that the Technology category boasts the highest profit margin at 17.4%, followed by Office Supplies at 17.0%, while Furniture lags at 2.5%. For regions, the West leads with a 14.9% margin, ahead of the East at 13.4% and Central at 7.9%. Itâ€™s exciting to see these high-margin areas, suggesting the business should focus resources on Technology products and the West region to boost profitability.
