Just tell pandas to try reading it with a different encoding like latin1 or ISO-8859-1, which usually works for Excel-generated or downloaded files.

errors='coerce' will turn any invalid dates into NaT (helps avoid crashes)

```python
#convert dat
df_Store_data['Order Date'] = pd.to_datetime(df_Store_data['Order Date'], errors='coerce')
```

```python
print(df_Store_data['Order Date'].dtype)
```

```python
#create Month_year
df_Store_data['month_year'] = df_Store_data['Order Date'].dt.to_period('M').astype(str)
```

```python
#Groupby month-year and calculate total sales
monthly_sales =  df_Store_data.groupby('month_year')['Sales'].sum().reset_index()
```

```python

plt.figure(figsize=(16,6))
plt.plot(monthly_sales['month_year'],monthly_sales['Sales'],marker='o', color='darkblue')
for i in range(len(monthly_sales)):
    plt.text(
        monthly_sales['month_year'][i],                     # x
        monthly_sales['Sales'][i] + 500,                    # y (a little above the point)
        f"{monthly_sales['Sales'][i]:,.0f}",                # formatted label (no decimals)
        ha='center', fontsize=10, rotation=45, color='black' # alignment, style
    )
plt.title('Monthly Sales Trend')
plt.xlabel('month_year')
plt.ylabel('Sales')
plt.xticks(rotation=45,fontsize=9)
plt.yticks(fontsize=9)
plt.grid(True)
plt.tight_layout()
plt.show()
```

1. What does the monthly/yearly sales trend look like? 

### Monthly Sales Trend – Insights
So I plotted the monthly sales data over a 4-year period, and a few interesting patterns really stood out:

- November is consistently the highest month every year (like in 2015, 2016, 2017). This is probably because of big shopping seasons maybe Black Friday, holiday offers, or year-end sales.

- February and April tend to be slower months, with noticeably lower sales. Makes sense could be post-holiday budget crunch or just naturally quiet months.

- Sales in 2017 are clearly higher overall compared to previous years. That suggests the business was growing steadily year by year.

Throughout the chart, there are regular ups and downs, which is pretty normal for any business. You can see those dips and recoveries happening often nothing too alarming, just part of the cycle.

### Conclusion 
I created this line chart to understand how sales have performed over time, month by month. It helps spot both the strong and weak periods. For example, sales are always high around November — likely due to seasonal promotions. On the flip side, months like February and April are usually slower. There's also a clear upward trend, especially in 2017, which points to business growth. These kinds of insights are super useful for planning — like when to stock more inventory, launch discounts, or run marketing campaigns.

```python
# Extract year from Order Date
df_Store_data['year'] = pd.to_datetime(df_Store_data['Order Date']).dt.year

# Group by year and sum sales
yearly_sales = df_Store_data.groupby('year')['Sales'].sum().reset_index()
```

```python
import matplotlib.pyplot as plt

# Plot
plt.figure(figsize=(10, 6))
plt.bar(yearly_sales['year'], yearly_sales['Sales'], color='darkblue', edgecolor='black')
plt.title('Yearly Sales Trend', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)

# sales values on top of each bar
for i in range(len(yearly_sales)):
    plt.text(
        yearly_sales['year'][i], 
        yearly_sales['Sales'][i] + 5000, 
        f"${yearly_sales['Sales'][i]:,.0f}", 
        ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group by Category
dec_category = dec15.groupby('Category')['Sales'].sum()
jan_category = jan16.groupby('Category')['Sales'].sum()

# Combine into single DataFrame
cat_compare = pd.DataFrame({'Dec 2015': dec_category, 'Jan 2016': jan_category})

# Plot
cat_compare.plot(kind='bar', figsize=(10,5))

# Adding labels to each bar
for i, category in enumerate(cat_compare.index):
    # Label for Dec 2015 bar
    plt.text(i - 0.2, cat_compare['Dec 2015'][category] + 500, 
             f"${cat_compare['Dec 2015'][category]:,.0f}", 
             ha='center', fontsize=10)
    # Label for Jan 2016 bar
    plt.text(i + 0.2, cat_compare['Jan 2016'][category] + 500, 
             f"${cat_compare['Jan 2016'][category]:,.0f}", 
             ha='center', fontsize=10)

plt.title('Sales by Category: Dec 2015 vs Jan 2016')
plt.ylabel('Sales')
plt.xticks(rotation=0)
plt.grid(True)
plt.tight_layout()
plt.show()
```

3. Are there any spikes or dips? Why? 

In the sales trend chart, I noticed a big drop in January 2016.
Sales fell from around ₹74,919 in December 2015 to just ₹18,542 in January 2016.
I checked the quantity of products sold — it also dropped a lot: from 1,203 units to 358.
So fewer products were sold overall.

Discount rates were almost the same, so it's not because of pricing.

But when I looked at product categories, I saw that sales in Technology dropped heavily, from ₹35,632 to just ₹5,620.

That’s a big reason for the overall dip.

Also, fewer orders came from all customer segments — especially Consumer, which had 185 orders in Dec and only 36 in Jan.

So the dip seems normal — probably due to year-end buying in December, and people slow down purchases in January.

### Insight: 
Sales dipped significantly in January 2016.

Total sales dropped from ₹74,919 in Dec 2015 to ₹18,542 in Jan 2016 (↓75%).

Quantity sold also declined (↓70%), indicating reduced customer activity.

Discount rates remained similar (~13%), so pricing was not the reason.

Major drop in the Technology category: ₹35,632 → ₹5,620.

All customer segments reduced orders, especially Consumer (↓80%).

All regions showed lower sales, with the West and East affected most.

### Conclusion:
The spike in December appears to be seasonal (possibly year-end or holiday promotions), followed by a natural slowdown in January.
No operational issue is visible — the dip aligns with typical post-holiday behavior.

```python
import matplotlib.pyplot as plt

# Group sales by Category
category_sales = df_Store_data.groupby('Category')['Sales'].sum().sort_values(ascending=False)

# Plotting
plt.figure(figsize=(8,5))
bars = plt.bar(category_sales.index, category_sales.values, color='teal')

# Add value labels above each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,  # X position (center of bar)
        height + 1000,                      # Y position (just above bar)
        f'{height:,.0f}',                   # Format with comma, no decimals
        ha='center', fontsize=9, color='black'
    )
plt.title('Total Sales by Product Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

### Insight
In this chart, we can clearly see that Furniture category has the highest sales, around ₹8.36 lakhs.
Then comes Technology with ₹7.42 lakhs, and Office Supplies is slightly lower at ₹7.19 lakhs.

I think one reason could be — furniture items are costlier than others, so even if quantity is less, the total value is high.

Overall, all three categories are performing well but furniture is clearly leading.”

```python
import pandas as pd
import matplotlib.pyplot as plt

# Grouping data by Region
region_group = df_Store_data.groupby('Region')[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False)

# Plotting
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Sales Bar Plot
ax1 = axs[0]
region_group['Sales'].plot(kind='bar', ax=ax1, color='teal')
ax1.set_title('Total Sales by Region')
ax1.set_ylabel('Sales')
ax1.set_xlabel('Region')
for bar in ax1.patches:
    ax1.annotate(f'{bar.get_height():,.0f}',
                 (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                 ha='center', va='bottom', fontsize=9)

# Profit Bar Plot
ax2 = axs[1]
region_group['Profit'].plot(kind='bar', ax=ax2, color='orange')
ax2.set_title('Total Profit by Region')
ax2.set_ylabel('Profit')
ax2.set_xlabel('Region')
for bar in ax2.patches:
    ax2.annotate(f'{bar.get_height():,.0f}',
                 (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                 ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

### Insight
When I looked at total sales and profit by region, I noticed something interesting.

- West region has the highest sales (around 725k) and also the highest profit (about 108k) — so that’s good, no issue there.

- East also has good sales (almost 679k) and decent profit (around 91k), so that looks balanced too.

- But the Central region is where I got confused a bit. Sales are quite high (501k), but profit is only around 39k, which is way less compared to West and East. That’s a sign that maybe something's off — like either cost is high or more discount. 

- South region has the lowest sales (391k), but even then, its profit (46k) is more than Central’s. Which again proves that Central region is not performing well in terms of profit, even though sales are good.

### Conclusion: 
Central region has high sales but low profit, so it needs attention. Maybe discounting too much, or cost structure is messed up.
